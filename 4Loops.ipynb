# Loops #

Often it is useful to repeat a process multiple times. To do so, we use loops. In Python, there are two types of loops, for loops and while loops.

* **For loops:** These loops are useful when you need to repeat a process a known number of times. 
* **While loops:** These loops are useful when you have to repeat a process an unknown number of times. The loop will repeat as long as some condition is met.

## For Loops ##

A for loop will repeat a process a set number of times. In Python, we can iterate a set number of times, or we can iterate over an iterable (such as a list, tuple, dict, or set). First, lets create a loop that iterates a set number of times. This loop will print the numbers 0, 1, and 2.

```python
for i in range(3):
    print(i)
```

Notice that the loop ends with a : and the next line needs an indention. If you do not have an indention or colon, the loop will not work. Also notice that the loop starts iterating at i=0, and stops at the value before 3. This is because Python indexes starting at 0, just like with lists. If we want to start and end the loop at different i values or increment by values besides 1, you can enter 3 numbers, using a similar listing convention to list slices. For example, if I want to print every other value of i between 7 and 1, in descending order I can use the following.

```python
for i in range(7,0,-2):
    print(i)
```

## While Loops ##
A while loop will iterate as long as the condition is true. The syntax below shows a while loop. Notice that the loop ends once the i value is no longer less than 3.

```python
i = 1
while i < 3:
    print(i)
    i += 0.5
```

If we want to break a loop prematurely, we can use the break command. Notice how the while loop stops after i reaches 2.

```python
i = 1
while i < 3:
    print(i)
    i += 0.5
    if i > 2:
        break
```

We can also return to the top of a loop prematurely with the continue command.

```python
i = 1
while i < 3:
    print(i)
    i += 0.5
    continue
    print("This string will not print.")
```

## Nested Loops ##

Another common use of loops in Python comes from nested loops. These are loops inside of loops. When calling these loops, you need to add another level of indention. The inside loop will iterate before the outside loop.

```python
for i in range(3):
    for j in range(2):
        print(i,j)
```
