## Variable Types ##

In Python, there are many prebuilt variable types. In the previous example, we created an int type. In Python, you do not need to explicitly declare the type of variable. It is automatically assumed. Here are some common variable types and their descriptions:

**int:** This is an integer. Integers are useful for discrete quantities and for iteration.

**bool:** This is a Boolean. It stores a truth value. Booleans are useful for if statements.

**float:** This is a float. Floats is useful for describing any real number. It can be used anywhere a real number can.

**str:** This is a string. It stores text. Strings are useful for any for storing anything with words.

**list:** This is a list. Lists are useful for storing objects in a particular order.

**tuple:** This is a tuple. Tuples function similarly to an ordered pair, but they can store more than just numbers.

**set:** This is a set. Sets store objects without a particular order. It is useful when order doesn't matter.

**dict:** This is a dictionary. Dictionaries store objects in an unordered structure with keys. They are useful for mappings.
    
This list or types is not exhaustive. We will talk about each type in more detail in future lessons.

## Floats ##

A float will function similar to a real number. In Python, there are multiple built in functions for math with real numbers. Here are some of them.

* **a + b** This adds a and b.
* **a - b** This subtracts b from a.
* **a * b** This multiplies a and b.
* **a / b** This divides a by b.
* **a ** b** This takes a to the power of b.
* **a // b** This divides a by b and rounds the quotient down to the nearest integer. This is called floor division.
* **a % b** This returns the remainder of a divided by b. This is called the modulo operation.

## Commenting ##

It is often to leave comments in your code, so you can understand what a particular variable means. In Python, a single line comment can be written with a #. Everything after the # in a given line will be passed over and ignored. This allows you to leave messages.



To leave a longer comment, use triple quotes on each side of the comment. Everything between the quotes will be passed over, even over multiple lines.
