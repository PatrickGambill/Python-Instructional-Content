# Conditionals #

A conditional is a statement like if or else. When the condition is met, the code inside of the if statement will run. Inside of an if statement, the indented code will run. Consider the example below.

```python
x = 2
y = 4

if x > 3:
    print("x")
if y > 3:
    print("y")
```

Notice that the condition for the first if statement is not met, but the second if statement's condition is met, so only the second if statement runs. Here are some common comparison operators and their syntax in Python.

* **==** This compares two values and returns True if they are equal.
* **!=** This compares two values and returns True if they are not equal.
* **<** This compares two values and returns True if the left term is less than the right term.
* **>** This compares two values and returns True if the left term is greater than the right term.
* **<=** This compares two values and returns True if the left term is less than or equal to the right term.
* **>=** This compares two values and returns True if the left term is greater than or equal to the right term.
* **in** This will return true if the element on the left is in the iterable on the right.

If you want a conditional to always evaluate as True or False, you can also use a Bool as the condition.

```python
if True:
    print("This will always print.")
if False:
    print("This will never print.")
```

## Boolean Logic ##

In Python, you can string together multiple conditions using and, or, and not. If you want two conditions to be true in order to run a statement, you can use and. If you want at least one of a set of conditions to be true, you can use or. If you want the code to run only when the condition is false, use not. Consider the example below.

```python
x = 1
y = 2

if x == 1 and y == 1:
    print("This will not print.")
if x == 1 or y == 1:
    print("This will print.")
if x == 1 and not y == 1:
    print("This will also print.")
```

## Else Statements ##

If you want to run code when the if statement is not met, you can use an else statement. An else statement will run if the if and elif statements above it are not met.

```python
x = 7

if x%2 == 0:
    print("x is even.")
else:
    print("x is odd.")
```

## Elif Statements ##
If you have multiple possible disjoint states, you can use elif statements. This is an abbreviation for else if and will only run if the if and elif statements above it are not met. Every elif and else statement after it will not run. 

```python
x = 7

if x < 3:
    print("This will not print.")
elif x < 6:
    print("This will not print either.")
elif x < 9:
    print("This will print.")
else:
    print("This will not print, since the elif condition above was already met.")
```
